Reverse the string

```python
def rev_string(str1):
  string1 = ''
  index = len(str1)
  while index >0:
    string1 += str1[index-1]
    index -=1
  return string1

strn = input()
print(rev_string(strn))
```

**Find maxium number between 3 given number**

```python
def max_of_two(x,y):
  if x>y:
    return x
  return y
def max_of_three(x,y,z):
  return max_of_two(x,max_of_two(y,z))

x = int(input())
y = int(input())
z = int(input())
print(max_of_three(x,y,z))
```
